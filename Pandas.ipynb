# Filtering
**Filtering** pada dasarnya memisahkan data berdasarkan suatu kondisi (*conditional*). Ini merupakan salah satu teknik analisis data, dimana kita ingin menampilkan data yang hanya memenuhi kondisi yang kita tentukan.

```python
people = {
    'first':['corey', 'jane', 'jhon'],
    'last':['schafer', 'doe', 'doe'],
    'email':['coreymschafer@example.com', 'janedoe@example.com', 'jhondoe@example.com']
    
}
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame(people)
```

```python
df1
```

```python
# Showing every record where the last name is Doe
filt = df1['last'] == 'doe'
filt
```

```python
df1[filt]
```

Dalam melakukan filtering kita menggunakan tanda logika seperti: '==', '>', '<', '>=', '<=', tujuannya adalah untuk membuat kondisi (*conditional*) yang menjadi *thumb of rule*. Pada contoh diatas kita membuat filter dimana baris yang akan ditampilkan harus berisi 'doe'pada kolom last, dan memasukkannya kedalam variabel filt, namun pandas menampilkan nilai boolean (True/False). <br>
nilai boolean (True/False) akan menjadi dasar bagi dataframe untuk memilih baris mana yang harus ditampilkan, jika nilai boolean = True maka baris tersebut, sedangkan Flase baris tidak akan ditampilkan. 

```python
df1.loc[filt, 'email']
```

Ada beberapa cara ntuk menampilkan hasil filtering dalam bentuk dataframe, yaitu:
1. menulis variabel dataset yang diikuti dengan baris kode filter, contoh: **df1[df1['last']=='doe']**.
2. menulis variabel dataset yang diikuti dengan variabel yang berisi baris kode filter, contoh: **df1[filt]**.
3. menggunakan "**.loc**", contoh: **df.loc[filt]**

Kita dapat menggunakan lebih dari 1 filter, caranya adalah dengan menambahkan 'AND(&)' dan 'OR(|)' diantara masing-masing filter (kondisi / *conditional*).

```python
# Multi filter using OR (|)
filt= (df1['last'] == 'schafer') | (df1['first'] == 'jhon')
```

```python
df1.loc[filt]
```

contoh kode:
> filt = (df1['last'] == 'schafer') | (df1['first'] == 'jhon')

Pada variabel filt diatas hanya akan menampilkan semua baris yang kolom awalnya corey atau yang kolom akhirnya jhon. hal ini dikarenakan sifat logika OR, dimana baris yang memenuhi minimal salah satu dari kedua kondisi filter akan ditampilkan.

```python
# Multi filter using AND (&)
filt = (df1['last'] == 'doe') & (df1['first'] == 'jhon')
```

```python
df1.loc[filt]
```

contoh kode:
> filt = (df1['last'] == 'doe') & (df1['first'] == 'jhon')

pada variabel filt diatas hanya akan menampilkan semua baris yang kolom awalnya jhon dan kolom akhirnya doe. hal ini disebebkan karena **sifat logika AND** yang **mengharuskan kedua kondisi terpenuhi**, sehingga semua baris yang tidak memenuhi kedua kondisi/kriteria tidak akan ditampilkan.

```python
# Multi filter using AND (&)
filt = (df1['last'] == 'doe') & (df1['first'] == 'jhon')
df1.loc[~filt]
```

**tanda negasi (~)** pada teknik filtering di pandas merupakan tanda NOT yang akan menampilkan baris nilai yang berkebalikan dengan kriteria filter. Pada kode diatas seharusnya hanya menampilkan baris yang memenuhi kedua kriteria filter, namun dikarenakan tanda negasi (~), hasil yang ditampilkan adalah nilai yang tidak memenuhi kedua kriteria filter. 

tanda negasi (~) dapat digunakan untuk membalik filter baik menggunakan logika AND ataupun logika OR

# Update Values
Update values atau memperbaharui nilai merupakan salah satu teknik yang harus dikuasai dalam melakukan data analisis terutama pada tahapan data cleaning. Pada tahapan pembersihan data (data cleaning) terkadang mengharuskan kita untuk memperbaiki data seperti: 
- Mengganti nilai
- Menyamakan font
- Memperbaiki typo (salah ketik)
- dll..<br>

Perbaikan dilakukan dengan mengganti nilai-nilai diatas dengan yang lebih sesuai</n>

```python
# Sample data
data = {
    "Buah (items)": ["Apel", "Pisang", "Jeruk", "Anggur"],
    "Hrga (price)": [5000, 3000, 7000, 10000],
    "Stok (quantity)": [10, 20, 15, 8]
}
```

```python
df = pd.DataFrame(data)
df
```

Salah penulisan nama kolom, pada contoh diatas terdapat kesalahan penulisan nama kolom **Hrga** yang seharusnya adalah **Harga**. Untuk memperbaikinya kita memerlukan fungsi bawaan pandas yaitu **.columns** diikuti dengan sebuah list berisi nama-nama kolom. berikut adalah contoh kodenya:

```python
df.columns = ['Buah (items)', 'Harga (price)', 'Stok (quantity)']
df
```

Ternyata setelah penulisan nama kolom yang diinginkan adalah uppercase. Untuk membuat nama kolom menjadi uppercase, kita dapat menggunakan sebuah _**list comprehension**_ dengan method string yaitu **.upper**, berikut adalah contoh kodenya:

```python
df.columns = [x.upper() for x in df.columns]
df
```

Selain **.upper** kita juga dapat menggunakan semua string method, seperti ''**.lower(); .capitalize(); dll**'' . Berikut adalah contoh baris kode menggunakan **.lower()**:

```python
df.columns = [x.lower() for x in df.columns]
df
```

Setelah dipikir-pikir ternyata nama kolom dengan penulisan lower case lah yang dipilih, namun dikarenakan nama kolom terdapat spasi yang menyebabkan *dot notation* menjadi tidak berfungsi. Untuk mengatasi hal ini kita memutuskan untuk mengganti "spasi" dengan "_", untuk melakukannya kita harus menggunakan fungsi **.str** yang diikuti dengan **.replace()**. Berikut adalah contoh kodenya:

```python
df.columns = df.columns.str.replace(' ', '_')
df
```

Selain teknik-teknik diatas, ada sebuah teknik yaitu **.rename** yang dapat digunakan untuk mengubah nama kolom, penjelasan lebih lengkap dapat pada documentation pandas.

## **Updating row's value**
untuk memperbaiki nilai pada baris dapat dilakukan dengan mencari baris yang ingin diubah terlebih dahulu, caranya ada 2 yaitu :
1. dengan teknik indexing seperti **.loc** atau **.iloc**
2. dengan teknik filtering

### __1. Teknik indexing seperti **.loc** atau **.iloc**__
Misalkan pada contoh data diatas, kita ingin memperbaharui harga Pisang menjadi 4000 dengan tenik indexing, yang menggunakan **.loc** dan meneruskan index baris yang diinginkan. seperti berikut:

```python
# selecting rows by index
df.loc[1] 
```

Setelah itu kita melakukan hal yang sama ketika mengubah nama kolom yang menggunakan **.columns** dan meneruskan sebuah list yang berisi nama kolom yang baru. Namun dalam konteks ini terdapat beberapa perubahan pada fungsi yang digunakan yaitu **.loc**, berikut adalah contoh kodenya:

```python
# add the new values in a list
df.loc[1] = ['Pisang', '4000', '10']
df
```

Kita juga dapat memanfaatkan salah satu kemampuan fungsi **.loc**, yaitu melakukan seleksi terhadap kolom menggunakan nama kolomnya, sehingga kita dapat langsung memilih nilai yang ingin diubah. <br>Misalkan ternyata Apel sudah dipakai sebanyak 5 buah dan kita ingin mengubahnya pada dataset, kita dapat melakukannya dengan menggunakan **.loc** diikuti dengan index baris pisang yaitu 0 dan index kolom stok yaitu stok_(quantity), berikut adalah contoh kodenya:

```python
# selecting a row based on index row and column's name, then adding a spesific new value
df.loc[0, 'stok_(quantity)'] = 5
df
```

### **2. Teknik Filtering**
Pada teknik ini, kita menggunakan teknik filtering untuk melakukan seleksi terlebih dahulu, lalu setelah itu baru memasukkan nilai baru. Misalkan pada contoh diatas kita ingin mengganti harga anggur menjadi 9500. berikut adalah contoh kodenya:

```python
filt = (df['buah_(items)']=='Anggur')
df[filt]['harga_(price)'] = 9500
```

error diatas disebabkan kita tidak menggunakan **.loc** ketika hendak mengisi nilai baru, berikut adalah baris kode yang seharusnya:

```python
filt = (df['buah_(items)']=='Anggur')
df.loc[filt, 'harga_(price)'] = 9500
df
```

**Intinya**: untuk melakukan update nilai baru pada dataset, kita memerlukan fungsi **.loc**, jadi kedua teknik diatas hanyalah teknik filtering yang dimodifikasi agar dapat melakukan update nilai pada dataset.
